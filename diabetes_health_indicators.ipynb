# 2- Construcci√≥n del modelo de regresi√≥n log√≠stica

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold, cross_val_score
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, RobustScaler
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.base import BaseEstimator, TransformerMixin
```

## 2.1- Auxiliares

```python
def train_val_test_split(df, rstate=42, shuffle=True, stratify=None):
    strat = df[stratify] if stratify else None
    train_set, test_set = train_test_split(
        df, test_size=0.4, random_state=rstate, shuffle=shuffle, stratify=strat)
    strat = test_set[stratify] if stratify else None
    val_set, test_set = train_test_split(
        test_set, test_size=0.5, random_state=rstate, shuffle=shuffle, stratify=strat)
    return (train_set, val_set, test_set)
```

```python
def remove_labels(df, label_name):
    X = df.drop(label_name, axis=1)
    y = df[label_name].copy()
    return (X, y)
```

```python
num_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", RobustScaler())
])
```

```python
# Transofrmador que prepara todo el conjunto de datos llamando pipelines y transformadores personalizados
class DataFramePreparer(BaseEstimator, TransformerMixin):

    def __init__(self):
        self._full_pipeline = None
        self._columns = None

    def fit(self, X, y=None):
        num_attribs = list(X.select_dtypes(exclude=['object']))
        cat_attribs = list(X.select_dtypes(include=['object']))
        self._full_pipeline = ColumnTransformer([
            ("num", num_pipeline, num_attribs),
            ("cat", CustomOneHotEncoder(), cat_attribs),
        ])
        self._full_pipeline.fit(X)
        self._columns = pd.get_dummies(X).columns
        return self

    def transform(self, X, y=None):
        X_copy = X.copy()
        X_prep = self._full_pipeline.transform(X_copy)
        return pd.DataFrame(X_prep,
                            columns=self._columns,
                            index=X_copy.index)
```

```python
class CustomOneHotEncoder(BaseEstimator, TransformerMixin):

    def __init__(self):
        self._oh = OneHotEncoder()
        self._columns = None

    def fit(self, X, y=None):
        X_cat = X.select_dtypes(include=['object'])
        self._columns = pd.get_dummies(X_cat).columns
        self._oh.fit(X_cat)
        return self

    def transform(self, X, y=None):
        X_copy = X.copy()
        X_cat = X_copy.select_dtypes(include=['object'])
        X_num = X_copy.select_dtypes(exclude=['object'])
        X_cat_oh = self._oh.transform(X_cat)
        X_cat_oh = pd.DataFrame(X_cat_oh.toarray(),
                                columns=self._columns,
                                index=X_copy.index)
        X_copy.drop(list(X_cat), axis=1, inplace=True)
        return X_copy.join(X_cat_oh)
```

## 2.2- Divisi√≥n del conjunto de datos

```python
train_set, val_set, test_set = train_val_test_split(
    df, rstate=42, shuffle=True, stratify="diagnosed_diabetes"
)
```

```python
X_train, y_train = remove_labels(train_set, 'diagnosed_diabetes')
X_val, y_val = remove_labels(val_set, 'diagnosed_diabetes')
X_test, y_test = remove_labels(test_set, 'diagnosed_diabetes')
```

## 2.3- Buscar hiperpar√°metros mas √≥ptimos para regresi√≥n log√≠stica

```python
# modelo base
pipeline = Pipeline([
    ("prep", DataFramePreparer()),
    ("clf", LogisticRegression(max_iter=5000, solver='saga', random_state=42))
])
```

```python
# b√∫squeda de hiperpar√°metros
param_grid = [
    {
        'clf__penalty': ['l1', 'l2', 'elasticnet', None],
        'clf__C': [0.01, 0.1, 1, 10, 100],
        'clf__solver': ['saga'],             # solver que soporta todos los penalties
        'clf__l1_ratio': [0.0, 0.5, 1.0]
    }
]
```

```python
# instanciar GridSearchCV
grid_search = GridSearchCV(
    pipeline,
    param_grid,
    cv=5,                        # validaci√≥n cruzada en 5 folds
    scoring='f1_weighted',       # m√©trica para evaluar
    return_train_score=True,
    n_jobs=-1
)
```

```python
grid_search.fit(X_train, y_train)

print(f'tiempo de b√∫squeda: {19} minutos {51} segundos')
print("Mejores hiperpar√°metros:", grid_search.best_params_)
print("Mejor modelo:", grid_search.best_estimator_)
print("Mejor score:", grid_search.best_score_)
```

## 2.4- Construir modelo mas √≥ptimo de clasificasi√≥n binaria

```python
# pipeline principal con modelo optimizado
pipe = Pipeline([
    ("prep", DataFramePreparer()),
    ("clf", LogisticRegression(
        C=0.1,
        l1_ratio=0.5,
        max_iter=5000,
        random_state=42,
        solver='saga'
    ))
])
```

```python
pipe.fit(X_train, y_train)
```

```python
# Evaluaci√≥n en datos de prueba
test_score = pipe.score(X_test, y_test)
print("Test score:", test_score)
```

```python
y_pred = pipe.predict(X_test)
```

## 2.5- M√©tricas del modelo

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score
import numpy as np
```

```python
acc = accuracy_score(y_test, y_pred)
print(f"Accuracy: {acc:.5f}")

print("\nReporte de clasificaci√≥n:\n")
print(classification_report(y_test, y_pred))

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")

plt.title("Matriz de confusi√≥n")
plt.xlabel("Predicciones")
plt.ylabel("Valores reales")

plt.show()
```

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

acc_scores = cross_val_score(pipe, X_train, y_train, cv=cv, scoring="accuracy")
print("Accuracy por fold:", acc_scores)
print("Media Accuracy:", np.mean(acc_scores))
print("Desviaci√≥n est√°ndar Accuracy:", np.std(acc_scores))

f1_macro_scores = cross_val_score(pipe, X_train, y_train, cv=cv, scoring="f1_macro")
print("\nF1-macro por fold:", f1_macro_scores)
print("Media F1-macro:", np.mean(f1_macro_scores))
print("Desviaci√≥n est√°ndar F1-macro:", np.std(f1_macro_scores))

f1_micro_scores = cross_val_score(pipe, X_train, y_train, cv=cv, scoring="f1_micro")
print("\nF1-micro por fold:", f1_micro_scores)
print("Media F1-micro:", np.mean(f1_micro_scores))
print("Desviaci√≥n est√°ndar F1-micro:", np.std(f1_micro_scores))
```
